```python
# The I/O module is used for importing the image
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from skimage import  io, data, color, filters
from skimage.color import rgb2gray
from skimage.transform import rescale, resize, downscale_local_mean
from skimage.exposure import rescale_intensity
```

```python
df = pd.read_csv('D:\Universidad\Data Science\proyecto 2\boneage-training-dataset.csv')
imagenes1 = []
#print(len(df['id_code;diagnosis']))
```

```python
#aqui cargamos la mitad del dataset de imagenes
for x in range(0, int(len(df['id_code;diagnosis'])/2)):
    name = df['id_code;diagnosis'][x][:-2] + ".png"
    #print(name)
    image = io.imread(name)
    imagenes1.append(image)
```

```python
print(int(len(df['id_code;diagnosis'])/2))
#print(len(df['id_code;diagnosis']))
```

```python
#probando como se ven las imagenes antes de procesarlas
#aqui se probo para ver el cambio del escalado y dimension
plt.imshow( resize(imagenes1[1000], (480, 640), anti_aliasing=True))
```

```python
#cambiar las dimensiones de las imagenes
for x in range(0, int(len(df['id_code;diagnosis'])/2)):
    imagenes1[x] = resize(imagenes1[x], (480, 640), anti_aliasing=True)
```

```python
#probando el cambio a escalas de grises
plt.imshow(color.rgb2gray(imagenes1[500]),cmap=plt.cm.gray)
```

```python
#volver a todas las imagenes en escalas de grises
for x in range(0, int(len(df['id_code;diagnosis'])/2)):
    imagenes1[x] = color.rgb2gray(imagenes1[x])
```

```python
#probando si todas las imagenes tiene el cambio a escala de grises
plt.imshow(imagenes1[500])
```

```python
#probando los diferentes escalados
plt.imshow(downscale_local_mean(imagenes1[2], (1, 1)))  
```

```python
#probando como seria la imagen con un filstro que resalte quizas lo mas importante en las imagenes.
#probando filtros
plt.imshow(rescale_intensity(1 - filters.sobel(imagenes1[500])), cmap=plt.cm.gray)  
#io.imsave('00000000.png',rescale_intensity(1 - filters.sobel(imagenes1[500])))
```

En este punto siento que lo mejor sera el convertir estas imagenes y hacer pruebas con ambos tipos.

```python
#convertiendo las imagenes en archivos ya estando estandarizadas, escaladas, redimesionadas y en escala de grises.
for x in range(0, int(len(df['id_code;diagnosis'])/2)):
    io.imsave(df['id_code;diagnosis'][x][:-2] +'.png',imagenes1[x])
```

```python
#convirtiendo las imagenes con filtro sobel
for x in range(0, int(len(df['id_code;diagnosis'])/2)):
    io.imsave(df['id_code;diagnosis'][x][:-2] +'.png',rescale_intensity(1 - filters.sobel(imagenes1[x])))
```

```python
#este es el codigo para procesar la otra mitad de las imagenes
df = pd.read_csv('train.csv')
imagenes2 = []
for x in range(1831,3662):
    name = df['id_code;diagnosis'][x][:-2] + ".png"
    #print(name)
    image = io.imread(name)
    imagenes2.append(image)
```

```python
#cambiar las dimensiones de las imagenes
for x in range(0, len(imagenes2)):
    imagenes2[x] = resize(imagenes2[x], (480, 640), anti_aliasing=True)
```

```python
#volver a todas las imagenes en escala de grises
for x in range(0, len(imagenes2)):
    imagenes2[x] = color.rgb2gray(imagenes2[x])
```

```python
#convertiendo las imagenes en archivos ya estando estandarizadas, escaladas, redimesionadas y en escala de grises.
for x in range(0, len(imagenes2)):
    io.imsave(df['id_code;diagnosis'][1831 + x][:-2] +'.png',imagenes2[x])
```

```python
#convirtiendo las imagenes con filtro sobel
for x in range(0, len(imagenes2)):
    io.imsave(df['id_code;diagnosis'][1831 + x][:-2] +'.png',rescale_intensity(1 - filters.sobel(imagenes2[x])))
```
