```python
# The I/O module is used for importing the image
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from PIL import Image
import scipy.stats as st
from pylab import *
```

```python
df = pd.read_csv('boneage-training-dataset.csv')
dimensiones = []
width = []
height = []
```

```python
for x in range(0, len(df)-1):
    im = np.array(Image.open('boneage-training-dataset/boneage-training-dataset/'+ str(df['id'][x])  + '.png'))
    h, w = im.shape
    width.append(w)
    height.append(h)
```

```python
for y in range(0, len(width)-1):
    dim = str(width[y]) + " x " + str(height[y])
    #print(dim)
    dimensiones.append(dim)
```

```python
tablaDim = pd.DataFrame({ 'dimensiones': dimensiones }) 
```

```python
count = tablaDim['dimensiones'].value_counts() 
bins = tablaDim['dimensiones'].unique()
print(bins)
```

```python
fig, axes = plt.subplots(figsize=(7,3), dpi=100)
plt.bar(bins, height=count)
plt.title('Histograma de frecuencia de dimensiones');
```

```python
plt.savefig('imagen.jpg',bbox_inches='tight', dpi=10050)
```
